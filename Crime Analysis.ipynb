# Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
crime=pd.read_csv('crime data.csv')
```

```python
df=crime.drop(labels=['Mocodes','Weapon Used Cd','Weapon Desc','Crm Cd 2','Crm Cd 3',
                      'Crm Cd 4','Cross Street'],axis=1)
df
```

```python
df.sort_values(by='Date Rptd')
```

```python
df.dtypes
```

```python
#Update Datatype
#df['Date Rptd']=pd.to_datetime(df['Date Rptd'],format='%Y-%m-%d')
#df['DATE OCC']=pd.to_datetime(df['DATE OCC'],format='%Y-%m-%d')

df['Date Rptd'] = pd.to_datetime(
    df['Date Rptd'],
    format='%Y %b %d %I:%M:%S %p'
)
df['DATE OCC'] = pd.to_datetime(
    df['DATE OCC'],
    format='%Y %b %d %I:%M:%S %p'
)

```

```python
df.info()
```

```python
df.describe(include=['float64','int64'])
```

```python
df['Vict Age'].value_counts()
```

```python
df_age=df[df['Vict Age']>0]
```

```python
df_age['Vict Age'].value_counts()
```

```python
df['year']=df['DATE OCC'].dt.year
df['month']=df['DATE OCC'].dt.month
```

```python
df
```

# Feature Understanding

```python
df['month'].value_counts().sort_index().plot()
```

```python
df['year'].value_counts().sort_index().plot(kind='bar')
```

```python
df = df[~((df['month'] == 12) & (df['year'] ==2023))]
```

```python
df.sort_values(by= 'DATE OCC')
```

```python
crime_over_time= df.groupby(['year','month']).size().plot()
crime_over_time
```

```python
plt.hist(df_age['Vict Age'],edgecolor='black',bins=15)
```

```python
crime_count_per_day=df.groupby(['DATE OCC','Crm Cd Desc']).size()
```

```python
top_20_crimes =crime_count_per_day.groupby('Crm Cd Desc').mean().reset_index(name='average_daily_count').sort_values(by='average_daily_count',ascending=False).head(20)
```

```python
top_20_crimes_sorted=top_20_crimes.sort_values(by='average_daily_count',ascending=True)
```

```python
plt.barh(top_20_crimes_sorted['Crm Cd Desc'],top_20_crimes_sorted['average_daily_count'])
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-ticks')
```

```python
plt.figure(figsize =(12,8))

#plot 1- Line Chart
plt.subplot(2,2,1)
crime_plot=df.groupby(['year','month']).size().plot()
crime_plot.set_xlabel('Year-Month',fontsize=10)
crime_plot.set_title('Crime Over Time')

#plot 2- Histogram
plt.subplot(2,2,2)
plt.hist(df_age['Vict Age'],edgecolor='black',bins=15)
plt.tight_layout()
plt.xlabel('Age',fontsize=10)
plt.title('Distribution of Victim Age')

#plot 3- Horizontal Bar
plt.figure()
plt.barh(top_20_crimes_sorted['Crm Cd Desc'],top_20_crimes_sorted['average_daily_count'])
plt.title('Top 20 Crimes')
```
